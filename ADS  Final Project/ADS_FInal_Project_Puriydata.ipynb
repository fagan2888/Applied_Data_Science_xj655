```python
import pandas as pd
import matplotlib as plt
import numpy as np
import pprint as pp
import re
```

```python
df_raw = pd.read_csv('llcp2017_formatted.csv')
```

```python
df_obv = df_raw.loc[:2000]
```

```python
df_obv.head(10)
```

```python
[print(i,df_obv[i].isnull().sum()) for i in df_obv.columns] #find how many null are in one column 
```

```python
dropcols = [i for i in df_obv.columns if df_obv[i].isnull().sum() > 1999]
dropcols
```

```python
len(dropcols)
```

```python
df_obv1 = df_obv.drop(dropcols,axis = 1)
```

```python
regMissing = re.compile(r'Missing')
tmp = df_obv1.replace(regMissing, np.nan).replace("Yes",1).replace("No",0)
#I replace the yes with 1, no with 0
tmp.head(10)
```

```python
print(list(tmp.columns))
```

```python
[i for i in tmp.columns if tmp[i].isnull().sum() > 1999]
```

```python
dropcols2 = [i for i in tmp.columns if tmp[i].isnull().sum() > 999]
df_obv2 = df_obv1.drop(dropcols2, axis = 1)
```
