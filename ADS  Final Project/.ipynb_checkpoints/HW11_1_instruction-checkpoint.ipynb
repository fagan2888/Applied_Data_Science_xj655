```python
from __future__ import print_function, division
import pandas as pd
import geopandas as gpd
import numpy as np
import os
import zipfile

try:
    from urllib import urlretrieve as urlretrieve
except ImportError:
    from urllib.request import urlretrieve as urlretrieve    
    
import pylab as pl
%pylab inline
```

```python
def getGeoDataFrameFromShpFileZipUrl(url):
    filenames = ! ls $PUIDATA | grep .shp
    urlretrieve(url, "file.zip")
    os.system('unzip -d $PUIDATA '+'file.zip')
    filenames2 = ! ls $PUIDATA | grep .shp
    shapeFile = [sf for sf in list(set(filenames2).difference(filenames)) if sf.split('.')[-1] == 'shp'][0]
    shapeFilePath = os.getenv('PUIDATA') + '/' + shapeFile
    return gpd.GeoDataFrame.from_file(shapeFilePath)
```

```python
shpurl = 'https://data.cityofnewyork.us/download/i8iw-xf4u/application%2Fzip'
nyczipsshp = getGeoDataFrameFromShpFileZipUrl(shpurl)
```

```python
parenturl = 'https://github.com/fedhere/PUI2018_fb55/raw/master/HW12_fb55/'
urls =[parenturl+'zbp'+"%02d" % (i,)+'totals.zip' for i in range(94,100)] + \
      [parenturl+'zbp'+"%02d" % (i,)+'totals.zip' for i in range(0,15)]

urls[:2]
```

```python
### This function downloads the pdf file, moves it into the PUIdata and reads it into a dataframe
def downloadZip(url):
    File = url.split('/')[-1]
    print("Downloading: ",File)
    urlretrieve(url, File)
    os.system('unzip -d $PUIDATA '+ File)
    os.system("mv " + File + ' ' + os.getenv("PUIDATA"))
```

```python
for url in urls:
    downloadZip(url)
```

```python
NYCzipcodes = nyczipsshp.ZIPCODE.values.astype(int)
print("length: ",len(NYCzipcodes))
NYCzipcodes[:5]
```

```python
df = None
for url in urls:
    fname = url.split('/')[-1].replace('zip','txt')
    year = fname[3:5]
    
    datadir = os.getenv("PUIDATA")+'/'
    currentdf = pd.read_csv(datadir+fname)
    currentdf.columns = [c.lower() for c in currentdf.columns]
    currentdf = currentdf[currentdf.zip.apply(lambda x: x in NYCzipcodes)]
    print("Shape of data in year ",year," is ",currentdf.shape)
    currentdf['year']= year
    if df is None:
        df = currentdf
    else:
        df = df.append(currentdf)
    print("Shape of accum data : ",df.shape)
```

Note on preprocessing: the preparation of these data should include the "standardization" of the time seris. 
*Always when you apply a moachine learning model you should standardize the data or whiten it.*
These operations may be complicated preprocessing operation, they can for example invoke PCA, and they are aimed at removing covariance in the feature space.

The simplest version of whitening *standardizes each feature*" from every column (feature) you subtract the mean and you divide it by the standard deviation.

But the time series preprocessing is different if you want to use each time stamp as feature: because you want to maintain the coherence of the data, i.e. you **do not want** to treat each feature as an independent feature! the pattenrs in the time series are what you are clustering on (time series goes up and down, creating a pattern) and those patterns would be lost by whitening.

*So when you cluster or otherwise use a time series in a model, using each timestamp as a feature, you want to standardize the observations, not the features.* 

**From each row of your table (observation, in this case each zipcode) subtract the mean and you divide the row by the standard deviation.**

```python
pivotdf = pd.pivot_table(estdf, values='est', index='zip',columns='year',aggfunc='sum')
```

```python
cbs = np.array(pivotdf)
```

## so these zips can be removed. 

```python
cleancsb = cbs[np.sum(np.isnan(cbs),axis=1)<=0]
```

```python
filteredzips = [z for z,row in zip(zips,cleancsb) if np.sum(np.isnan(row)) <= 0]
len(filteredzips)
```

```python
#cleaning all missing values
cleancsb.shape
```

```python
#then the standardization will look like:
vals = ((cleancsb.T - cleancsb.mean(1)) / cleancsb.std(1)).T
vals.shape
```

vals.mean(axis=1) should return 0 (or something very small) everywhere

vals.mean(axis=1).shape should be (260,)

vals.std(axis=1) should return 1 everywhere

vals.std(axis=1).shape

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
nc = 5
```

```python
#cluster with kmeans
km = KMeans(n_clusters=nc, random_state=444).fit(vals)
```

```python
#each observation in the km object has a label: 
#that is the cluster assigned to that observation
km.labels_.shape
```

```python
set(km.labels_)
```

```python
len(km.cluster_centers_[0])
```

### The map of the clusters may look something like this

```python
labeldf = pd.DataFrame({ 'zip':filteredzips, 'label':km.labels_ })
```

```python
nyczipsshp['ZIPCODE'] = nyczipsshp.ZIPCODE.astype(int)
nyczipsshp = nyczipsshp.merge(labeldf, left_on = 'ZIPCODE', right_on = 'zip', how = 'inner' )
nyczipsshp.shape
```

```python
import choroplethNYC as cp
cp.choroplethNYC(nyczipsshp, column='label', scheme='Equal_Interval',k=nc, cmap="bone",  edgecolor='black', lw=1);
```

```python
labels = km.labels_
for i in range(nc):
    fig=pl.figure(figsize=(10,3))
    ax=fig.add_subplot(111)
    idx = labels == i
    data = vals[idx]
    ax.plot(data.T)
    ax.plot(km.cluster_centers_[i], lw=15)
    #ax.plot(np.mean(data, axis=0), lw=15)
    
```
