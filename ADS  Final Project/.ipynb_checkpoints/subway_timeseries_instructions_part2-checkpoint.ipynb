# 4. Fourier transform
from the numpy package you will nees only the 
np.fft.rfft() to take the fourier transform

and

np.fft.rfftfreq() to create a meaningful x axis 

`````python
## Since we are only interested in subway stations I will aggregate data by stations
mtaStation = np.nansum(mta, axis=1)
mtaStation.shape
`````

`````python
f = (np.fft.rfft(mtaStation[0]).real)**2
`````

`````python
N= mtaStation.shape[1]
sample_rate=1

fig=pl.figure(figsize=(15,5))
ax=fig.add_subplot(111)
ax.plot(np.fft.rfftfreq(N, 1.0/sample_rate), (f), 'o', ms=20)
ax.plot(np.fft.rfftfreq(N, 1.0/sample_rate), (f), '-')

ax.set_xticklabels([ "%.2f"%(fx) for fx in ax.get_xticks()], fontsize=20)
pl.xlabel("frequency (1/seconds - Hz)", fontsize=20)
pl.ylabel("Power", fontsize=20)
pl.title("Power Spectrum", fontsize=20);
`````

`````python
## From professors notebook:https://github.com/fedhere/UInotebooks/blob/master/timeseries/fourier.ipynb
print (" The first two frequency (slowest) caculated are \n1: infinity (1.0/0) ",
       "and \n2:the frequency corresponding to the entire duration of the series = N*d,"
       " \nand they should be removed because they are uninteresting\n",
       "freq: ", np.fft.rfftfreq(N, 1.0)[0], "= period: ", 1.0/np.fft.rfftfreq(N, 1.0)[0], "and\n",
       "freq: ", np.fft.rfftfreq(N, 1.0)[1], "= period: every %.1f"%(1.0/np.fft.rfftfreq(N, 1.0)[1]), "weeks\n\n")
`````

`````python
fig=pl.figure(figsize=(15,5))
ax=fig.add_subplot(111)
ax.plot(np.fft.rfftfreq(N, 1.0)[4:], (f[4:]), 'o', ms=20)
ax.plot(np.fft.rfftfreq(N, 1.0)[4:], (f[4:]), '-')

# changing the labels on the x ticks to see the period instead of frequency
ax.set_xticklabels([ "%s"%(1/fx) for fx in ax.get_xticks()], fontsize=20)
pl.xlabel("period (week)", fontsize=20)
pl.ylabel("Power", fontsize=20)
pl.title("Trash pickup power spectrum", fontsize=20);
pl.vlines(1/52, 0, 1e12, linestyle='--')
`````

`````python
flattenedmta = mta.reshape(mta.shape[0]*mta.shape[1],mta.shape[2])
flattenedmta.shape

# sanity check
mta[1][0][0] == flattenedmta[23][0]

flatRiderType = np.tile(ridetype, mta.shape[0])
flatRiderType.shape

# sanity check
flatRiderType[0] == flatRiderType[2*23]

meanstation = np.nanmean(mta, axis=(1,2))
print(meanstation.shape)

flatMeanStation = np.repeat(meanstation,mta.shape[1])
flatMeanStation.shape

# sanity check
flatMeanStation[0] == flatMeanStation[22]  and flatMeanStation[0] != flatMeanStation[23]

stdstation = np.nanstd(mta, axis=(1,2))
print(stdstation.shape)
flatstdStation = np.repeat(stdstation,mta.shape[1])
flatstdStation.shape

feature1 = np.divide(np.nanmean(flattenedmta,axis=1), flatMeanStation)
feature1.shape

print("Missing values feature 1: ", np.sum(np.isnan(feature1)) )

feature2 = np.divide(np.nanstd(flattenedmta,axis=1), flatstdStation)
feature2.shape

print("Missing values feature 1: ", np.sum(np.isnan(feature2)) )

def getRegCoeff(Y):
    if Y.std() > 0:
        X = sm.add_constant(np.array(range(len(Y))))
        mod = sm.OLS(Y,X).fit()
        return mod.params[0],mod.params[1]
    else : 
        return (np.nan,np.nan)

feature34 = [getRegCoeff(flattenedmta[i]) for i in range(flattenedmta.shape[0])]

feature3 = [i for i,s in feature34]
feature4 = [s for i,s in feature34]

print("Missing values feature 3: ", np.sum(np.isnan(feature3)) )
print("Missing values feature 4: ", np.sum(np.isnan(feature4)) )

Y = flatRiderType

Xzip = [(f1,f2,f3,f4,y) for (f1,f2,f3,f4,y) in zip(feature1,feature2,feature3,feature4,Y) if \
       not (np.isnan(f1) or np.isnan(f2) or np.isnan(f3) or np.isnan(f4)) ]

len(Xzip)

X = np.array(Xzip)[:,0:4]
X = np.array(X,np.dtype('float'))
X.shape

Y = np.array(Xzip)[:,4]
Y.shape
`````

`````python
X.shape
`````

`````python
meanX = np.mean(X,axis=0)
stdX = np.std(X,axis=0)
X = np.divide(X-meanX,stdX)
`````

`````python
#from scipy.cluster.vq import kmeans2, whiten
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
`````

`````python
np.random.seed(111)
nk= 5
cluster_centroids, closest_centroids = kmeans2(X, nk, iter=50)
`````

`````python
# k means determine k
np.random.seed(111)
distortions = []
K = [10,23,30,40,50]
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
`````

`````python
# Plot the elbow
fig=pl.figure(figsize=(15,5))
pl.plot(K, distortions, 'bx-')
pl.xlabel('k')
pl.ylabel('Distortion')
_=pl.title('The Elbow Method showing the optimal k')
`````

### We kind of observe the elbow twist after 23 so data is being clustered by ticket types.

`````python
# k means determine k
np.random.seed(111)
distortions = []
K = [500,600,700,800]
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
`````

`````python
# Plot the elbow
fig=pl.figure(figsize=(15,5))
pl.plot(K, distortions, 'bx-')
pl.xlabel('k')
pl.ylabel('Distortion')
_=pl.title('The Elbow Method showing the optimal k')
`````

### We see a very slight bend at 600 unlike at 23 so the elbow test indicates clustering by ticket type
